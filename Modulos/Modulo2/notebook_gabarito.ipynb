# Resolução dos Exercícios com Pandas

Dataset: dataset_vendas.csv

## Carregar os dados

```python
import pandas as pd

df = pd.read_csv('dataset_desempenho_escolar.csv')
df.head()
```

## 1. Criar um DataFrame maior com base no dataset original

```python
# Concatenando o dataset com ele mesmo para aumentar o tamanho
df_maior = pd.concat([df, df], ignore_index=True)
df_maior.shape
df = pd.DataFrame(df_maior)
df
```

## 2. Remover valores negativos

```python
# Removendo valores negativos
df_limpo = df_maior[(df_maior.select_dtypes(include=['number']) >= 0).all(axis=1)]
df_limpo.head()
```

## 3. Criar coluna de multiplicação (nota x frequencia)

```python
# Supondo que existam colunas 'Nota' e 'Frequencia'
df_limpo['multiplicacao_calc'] = df_limpo['nota'] * df_limpo['frequencia']
df_limpo.head()
```

## 4. 5 notas mais altas

```python
# Agrupando por nota
df_limpo[df_limpo['nota'] == df_limpo['nota'].max()]
df_limpo.sort_values(by='nota', ascending=False).head(5)
```

## 5. Média de notas

```python
media_notas = df_limpo['nota'].mean()
media_notas
```
