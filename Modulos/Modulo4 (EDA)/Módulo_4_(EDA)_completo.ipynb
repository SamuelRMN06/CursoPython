**3.2 Módulo 4 (EDA) completo**

O Módulo 4 foi dedicado à Análise Exploratória de Dados (EDA), etapa responsável por examinar o dataset Brazilian Cities e compreender o comportamento das principais variáveis presentes na base. O objetivo foi identificar padrões, tendências, relações entre indicadores e possíveis inconsistências nos dados, preparando uma base sólida para a construção do projeto final.

Durante essa fase, foram analisadas informações relacionadas ao Índice de Desenvolvimento Humano Municipal (IDHM), PIB per capita, classificação territorial dos municípios e composição econômica. A EDA permitiu transformar dados brutos em informações organizadas e relevantes para apoiar as próximas etapas do trabalho.

**3.2.1 Notebook de EDA estruturado**

O notebook foi desenvolvido de forma organizada e progressiva, seguindo uma sequência lógica que facilitou tanto a execução quanto a interpretação da análise.

Inicialmente, foram importadas bibliotecas essenciais para manipulação e visualização de dados, como Pandas, Matplotlib e Seaborn. Em seguida, foi realizado o carregamento do dataset e uma inspeção inicial da base, com verificação de colunas, tipos de dados e estatísticas descritivas.

Posteriormente, o notebook foi dividido em blocos temáticos, cada um voltado para responder uma pergunta específica da análise. Foram explorados temas como distribuição do IDHM, ranking dos estados com melhor desempenho, relação entre renda e desenvolvimento humano, perfil rural e urbano dos municípios e setores econômicos predominantes.

Essa estrutura proporcionou clareza metodológica, boa organização visual e alinhamento com os objetivos definidos pelo grupo.

**3.2.2 Análise guiada**

A exploração dos dados foi conduzida a partir de perguntas previamente definidas, o que garantiu foco e objetividade ao longo de todo o processo.

Cada etapa da análise buscou responder questões relevantes sobre os municípios brasileiros, como a distribuição do desenvolvimento humano, diferenças entre estados, influência do PIB per capita, características territoriais e participação econômica dos setores produtivos.

Essa abordagem guiada contribuiu para evitar análises dispersas, tornando o processo mais estratégico e diretamente conectado aos objetivos do projeto final.

**3.2.3 Storytelling com dados**

A análise foi organizada em uma sequência lógica de interpretação, utilizando o conceito de storytelling com dados para facilitar a compreensão dos resultados.

O processo iniciou com uma visão geral do IDHM nos municípios brasileiros, permitindo entender o cenário nacional. Em seguida, foram apresentadas comparações entre estados e posteriormente relações entre desenvolvimento humano e fatores econômicos.

Na sequência, a análise abordou a divisão territorial entre municípios rurais e urbanos, finalizando com a participação dos setores econômicos na composição do PIB municipal.

Essa estrutura narrativa permitiu transformar informações isoladas em uma leitura coerente e progressiva sobre a realidade socioeconômica brasileira.

**3.2.4 Conclusões bem escritas**

A Análise Exploratória de Dados permitiu identificar pontos relevantes sobre os municípios brasileiros.

Observou-se que grande parte dos municípios apresenta níveis intermediários de IDHM, enquanto poucos alcançam índices mais elevados. Também foi possível verificar que estados das regiões Sul e Sudeste tendem a apresentar melhores médias de desenvolvimento humano.

Além disso, percebeu-se uma relação positiva entre PIB per capita e IDHM, indicando que renda influencia o desenvolvimento, embora não seja o único fator determinante. Também foi constatada forte presença de municípios com características rurais e predominância do setor de serviços na economia local.

Essas conclusões forneceram base analítica consistente para as visualizações e insights apresentados na etapa final do projeto.
