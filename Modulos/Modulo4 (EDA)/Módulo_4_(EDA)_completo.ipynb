## Primeira Versão do Projeto EDA

### Dataset Escolhido
**Brazilian Cities Dataset** — Dados socioeconômicos de 5.578 municípios brasileiros.

### Problema Formulado
> *"O que determina o nível de desenvolvimento humano (IDHM) de um município brasileiro, e como ele varia entre regiões, estados e categorias urbanas?"*

### Perguntas Norteadoras

1. **Distribuição geográfica:** Como o IDHM varia entre as cinco regiões do Brasil? Quais estados têm os maiores e menores índices?

2. **Fatores associados:** Quais variáveis (PIB per capita, densidade demográfica, infraestrutura) têm maior correlação com o IDHM?

3. **Desigualdade:** Qual o grau de desigualdade no IDHM entre municípios do mesmo estado? Existe uma "cauda longa" de cidades muito pobres?

4. **Urbanização:** Cidades mais urbanizadas têm sistematicamente IDHM mais alto? Existe exceção relevante?

5. **Componentes:** Qual dos três componentes do IDHM (Renda, Longevidade, Educação) é o "gargalo" em cada região?

6. **Tamanho e desenvolvimento:** Municípios maiores são necessariamente mais desenvolvidos? Existem municípios pequenos com alto IDHM?



**3.2 Módulo 4 (EDA) completo**

O Módulo 4 foi desenvolvido com foco na Análise Exploratória de Dados (EDA), com o objetivo de investigar os fatores que influenciam o Índice de Desenvolvimento Humano Municipal (IDHM) no Brasil. A análise foi conduzida de forma estruturada, utilizando estatísticas descritivas e visualizações para identificar padrões, relações e desigualdades entre municípios, regiões e estados.

**3.2.1 Notebook de EDA estruturado**

 O notebook foi organizado de forma lógica e progressiva, iniciando com a apresentação do dataset e do problema formulado. Em seguida, foram incluídas etapas de compreensão e preparação dos dados, garantindo a consistência das informações analisadas.
Posteriormente, o notebook foi dividido em seções baseadas nas perguntas norteadoras, abordando temas como distribuição geográfica do IDHM, fatores associados, desigualdade, urbanização, componentes do índice e relação entre tamanho populacional e desenvolvimento. Cada seção contém análises acompanhadas de explicações, facilitando a leitura e interpretação dos resultados.

**3.2.2 Análise guiada**  

A análise foi conduzida a partir das perguntas norteadoras definidas previamente, o que garantiu foco e objetividade ao longo de todo o processo. Cada etapa da EDA teve como objetivo responder questões específicas, como a variação do IDHM entre regiões, a influência de fatores socioeconômicos e a existência de desigualdades entre municípios.
Essa abordagem evitou análises superficiais ou desconectadas, permitindo uma investigação mais aprofundada e orientada a gerar insights relevantes sobre o desenvolvimento humano no Brasil.

**3.2.3 Storytelling com dados**

Os resultados foram organizados em uma narrativa lógica, utilizando storytelling com dados para facilitar a compreensão. A análise iniciou com uma visão geral da distribuição do IDHM no Brasil, seguida por comparações regionais e estaduais.
Em seguida, foram exploradas as relações entre o IDHM e variáveis socioeconômicas, destacando fatores mais influentes. Posteriormente, a análise abordou desigualdades internas, urbanização e os componentes do IDHM, identificando possíveis gargalos em diferentes regiões.
Essa construção narrativa permitiu conectar os dados às perguntas iniciais, tornando os resultados mais claros, interpretáveis e relevantes.

**3.2.4 Conclusões bem escritas**  

As conclusões foram elaboradas com base nos resultados obtidos ao longo da EDA, respondendo diretamente ao problema proposto. Foram identificadas diferenças significativas no IDHM entre regiões brasileiras, além de relações importantes com fatores como renda e urbanização.
Também foram observados níveis de desigualdade dentro de estados, evidenciando a existência de municípios com baixo desenvolvimento em contraste com outros mais desenvolvidos. A análise dos componentes do IDHM permitiu identificar gargalos específicos, como educação ou renda, dependendo da região.
Por fim, verificou-se que o tamanho do município não é, por si só, um determinante absoluto do desenvolvimento, existindo exceções relevantes que reforçam a complexidade do fenômeno analisado.
