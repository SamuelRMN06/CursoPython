## 1. Importação de Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 2. Carregamento dos Dados

```python
df = pd.read_csv('brazilian_city.csv')
df.head()
```

## Pergunta 1 — Como o IDHM está distribuído entre os municípios?

```python
plt.figure()
plt.hist(df['IDHM'].dropna(), bins=30, color='steelblue', edgecolor='white')
plt.title('Distribuição do IDHM nos Municípios Brasileiros')
plt.xlabel('IDHM')
plt.ylabel('Número de municípios')
plt.axvline(df['IDHM'].mean(), color='red', linestyle='--', label=f'Média: {df["IDHM"].mean():.3f}')
plt.legend()
plt.show()
```

**Insight:** A maioria dos municípios tem IDHM entre 0,6 e 0,75, com poucos atingindo valores acima de 0,8. A distribuição é levemente assimétrica à esquerda, indicando que ainda há muitas cidades com desenvolvimento abaixo da média.

---
## Pergunta 2 — Quais os 10 estados com maior IDHM médio?

```python
top_estados = df.groupby('STATE')['IDHM'].mean().sort_values(ascending=False).head(10)

plt.figure()
top_estados.plot(kind='bar', color='steelblue')
plt.title('Top 10 Estados com Maior IDHM Médio')
plt.xlabel('Estado')
plt.ylabel('IDHM Médio')
plt.xticks(rotation=45)
plt.ylim(0.6, 0.85)
plt.show()
```

**Insight:** Os estados do Sul (SC, RS, PR) e Sudeste (SP, RJ) dominam o ranking, refletindo melhor infraestrutura, educação e renda nessas regiões.

---
## Pergunta 3 — Existe relação entre PIB per capita e IDHM?

```python
df_scatter = df[df['GDP_CAPITA'] < 150000].dropna(subset=['GDP_CAPITA', 'IDHM'])

plt.figure()
sns.scatterplot(data=df_scatter, x='GDP_CAPITA', y='IDHM', alpha=0.4)
plt.title('PIB per capita × IDHM')
plt.xlabel('PIB per capita (R$)')
plt.ylabel('IDHM')
plt.show()
```

**Insight:** Há uma correlação positiva entre PIB per capita e IDHM, porém não é perfeita — municípios com renda alta nem sempre têm IDHM alto, o que sugere que educação e saúde também são fatores determinantes.

---
## Pergunta 4 — Como os municípios se dividem entre área rural e urbana?

```python
contagem = df['RURAL_URBAN'].value_counts()

plt.figure()
contagem.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proporção de Municípios por Tipo de Área')
plt.ylabel('')
plt.show()
```

**Insight:** A maioria dos municípios brasileiros é classificada como Rural Adjacente, o que mostra que o Brasil ainda tem uma grande parte do território com características semi-rurais, apesar da alta urbanização nas grandes cidades.

---
## Pergunta 5 — Qual setor econômico domina o PIB dos municípios?

```python
setores = {
    'Agropecuária': df['GVA_AGROPEC'].sum(),
    'Indústria':    df['GVA_INDUSTRY'].sum(),
    'Serviços':     df['GVA_SERVICES'].sum(),
    'Público':      df['GVA_PUBLIC'].sum()
}

plt.figure()
plt.bar(setores.keys(), [v / 1e9 for v in setores.values()], color=['green', 'steelblue', 'orange', 'purple'])
plt.title('Soma do Valor Adicionado por Setor Econômico')
plt.xlabel('Setor')
plt.ylabel('Valor (R$ bilhões)')
plt.show()
```

**Insight:** O setor de Serviços é o maior contribuinte do PIB dos municípios brasileiros, seguido pelo setor Público e pela Indústria. A Agropecuária, apesar de importante, representa a menor fatia do valor adicionado total.
