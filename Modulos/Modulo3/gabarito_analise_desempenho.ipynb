# 📊 Exercícios de Análise de Dados: Desempenho Escolar

Este notebook apresenta 5 questões práticas para explorar o seu conjunto de dados de desempenho escolar utilizando as bibliotecas **Matplotlib** e **Seaborn** para gerar visualizações modernas e informativas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração visual moderna
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Carregando os dados
# Certifique-se de que o arquivo 'dataset_desempenho_escolar.csv' está no mesmo diretório
try:
    df = pd.read_csv("dataset_desempenho_escolar.csv")
    print("Dataset carregado com sucesso!")
    print(df.head())
except FileNotFoundError:
    print("Erro: O arquivo 'dataset_desempenho_escolar.csv' não foi encontrado.")
```

### 1. Distribuição de Desempenho
**Questão:** Como as notas estão distribuídas na turma? Existe uma concentração em notas altas ou baixas?

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['nota'], kde=True, color='skyblue', bins=6)
plt.title('Distribuição das Notas da Turma', fontsize=14)
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()
```

### 2. Ranking de Alunos
**Questão:** Quem são os alunos com melhor desempenho e como eles se comparam à média?

```python
plt.figure(figsize=(10, 6))
media_nota = df['nota'].mean()
df_sorted = df.sort_values('nota', ascending=False)

sns.barplot(data=df_sorted, x='nome', y='nota', palette='magma')
plt.axhline(media_nota, color='red', linestyle='--', label=f'Média: {media_nota:.1f}')

plt.title('Ranking de Notas por Aluno', fontsize=14)
plt.legend()
plt.show()
```

### 3. Relação Frequência vs. Nota
**Questão:** Existe uma correlação visual entre a presença em aula e a nota final?

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='frequencia', y='nota', scatter_kws={'s':100}, line_kws={'color':'orange'})
plt.title('Influência da Frequência no Desempenho Escolar', fontsize=14)
plt.xlabel('Frequência (%)')
plt.ylabel('Nota Final')
plt.show()
```

### 4. Análise de Engajamento
**Questão:** Qual a dispersão da frequência dos alunos? Existem alunos com comportamento muito distante da maioria?

```python
plt.figure(figsize=(6, 8))
sns.boxplot(y=df['frequencia'], color='lightgreen', width=0.4)
sns.swarmplot(y=df['frequencia'], color='black', alpha=0.5)
plt.title('Dispersão da Frequência Escolar', fontsize=14)
plt.ylabel('Frequência (%)')
plt.show()
```

### 5. Matriz de Correlação
**Questão:** Qual a força da relação numérica entre todas as variáveis do dataset?

```python
plt.figure(figsize=(8, 5))
# Selecionando apenas colunas numéricas
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor: Correlação entre Variáveis', fontsize=14)
plt.show()
```
