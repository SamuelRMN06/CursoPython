## 1. Importação de Bibliotecas
Aqui carregamos as bibliotecas principais:
- `pandas`: manipulação de dados
- `matplotlib`: criação de gráficos
- `seaborn`: gráficos mais bonitos e estatísticos


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
````

## 2. Carregamento dos Dados
Agora vamos carregar o dataset.
Certifique-se de que o arquivo CSV está no mesmo diretório do notebook.


````python
df = pd.read_csv('dataset_desempenho_escolar.csv')
df.head()
````

## 3. Análise Exploratória
Antes de criar gráficos, precisamos entender os dados:
- `info()`: mostra tipos de dados
- `describe()`: estatísticas básicas


````python
df.info()
df.describe()
````

## 4. Matplotlib - Fundamentos
Vamos criar um gráfico simples:
👉 Total de vendas por produto

Passos:
1. Agrupar os dados
2. Criar gráfico de barras
3. Adicionar título e rótulos


````python
vendas_produto = df.groupby('nota')['frequencia'].sum()

plt.figure()
vendas_produto.plot(kind='bar')
plt.title('Total de Notas por Frequência')
plt.xlabel('Nota')
plt.ylabel('Frenquência')
plt.xticks(rotation=45)
plt.show()
````

## 5. Seaborn - Gráficos Estatísticos
Agora vamos usar o Seaborn para análises mais profundas.

### 📦 Boxplot
Mostra a distribuição dos dados e possíveis outliers.


````python
plt.figure()
sns.boxplot(x='nota', y='frequencia', data=df)
plt.title('Distribuição de Notas por Frequência')
plt.xticks(rotation=45)
plt.show()
````

### Heatmap (Correlação)
Mostra a relação entre variáveis numéricas.


````python
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de Correlação')
plt.show()
````

### Regressão Linear
Ajuda a identificar tendências entre variáveis.


````python
plt.figure()
sns.regplot(x='nota', y='frequencia', data=df)
plt.title('Relação entre Nota e Frequência')
plt.show()
````

## 7. Boas Práticas
- Não polua o gráfico com informação demais
- Use cores com significado
- Sempre rotule e coloque título
- Escolha o gráfico certo para o tipo de dado

Lembre-se: o objetivo não é só mostrar dados, mas contar uma história com eles.
